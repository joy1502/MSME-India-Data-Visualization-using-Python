```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rcParams["figure.figsize"] = (16, 7)
```

```python
state = ["Uttar pradesh", "West bengal", "Tamil Nadu", "Maharastra", "Karnataka", "Bihar", "Andhra pradesh", "Gujarat", "Rajasthan", "Madhya pradesh", "Total of above ten states", "Other state/UTs"]
number = [89.99, 88.67, 49.48, 47.78, 38.34, 34.46, 33.87, 33.16, 26.87, 26.74, 469.36, 164.52]

c = "black", "teal", "gray"
fig=plt.figure(figsize=(7,2))
plt.bar(state,number, color = c)
plt.xlabel('states',fontsize=20)
plt.ylabel('number',fontsize=20)
plt.title("Number of estimated MSMEs", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xticks(rotation=60)
plt.show()
```

```python
year = ["2014-15", "2015-16", "2016-17", "2017-18", "2018-19"]
allindiaGDP = [29.34, 29.48, 29.25, 29.75, 30.27]

c = "gray", "black"
fig=plt.figure(figsize=(5,3))
plt.bar(year,allindiaGDP, color = c)
plt.xlabel('states',fontsize=5)
plt.ylabel('number',fontsize=5)
plt.title("Number of estimated MSMEs", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```
